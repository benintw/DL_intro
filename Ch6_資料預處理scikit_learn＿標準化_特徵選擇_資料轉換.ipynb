### StandardScaler

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("/content/drive/MyDrive/Python 資料科學自學聖經/data_csv/汽車車型資料檔.csv")
df
```

```python
df.columns
```

```python
std = StandardScaler() 
df1 = df[['哩/加侖', '氣缸', '排氣量', '馬力', '重量', '加速時間', '年份']] # StandardScaler 只能針對數值欄位做標準化, 所以省去'名稱 '
arr = std.fit_transform(df1)
df[['哩/加侖', '氣缸', '排氣量', '馬力', '重量', '加速時間', '年份']] = arr
df
```

### LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
label1 = LabelEncoder()
df = pd.read_csv("/content/drive/MyDrive/Python 資料科學自學聖經/data_csv/客戶聯絡狀況資料檔.csv")
df["工作"] = label1.fit_transform(df["工作"])
df
```

```python
label1.classes_
```

```python
# 將轉換後的數值還原為特徵值
df["工作"] = label1.inverse_transform(df["工作"])
df
```
